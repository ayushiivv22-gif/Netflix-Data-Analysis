```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix.csv")
df.head()
```

```python
# BUSINESS RELATED ORIENTED QUESTIONS.
#(i) What is the ratio of Movies vs TV Shows on Netflix?
```

```python
type_counts = df['type'].value_counts()

type_counts.plot(kind='pie', autopct='%1.1f%%', colors=['pink','blue'])
plt.title("Movies vs TV Shows")
plt.ylabel('')
plt.show()
```

```python
#INSIGHTS
# Movies dominate the platform compared to TV Shows, indicating higher investment in short-form content.
# TV Shows still hold a significant share, showing growing demand for long-form storytelling.
```

```python
#(ii)  Which genres are most popular on Netflix globally?
```

```python
#code
genres = df['listed_in'].str.split(', ', expand=True).stack()
genres.value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title("Top Genres")
plt.xticks(rotation=45)
plt.show()
```

```python
#INSIGHTS
# Drama and International genres are usually among the most common, showing global audience preference.
# Diverse genre availability suggests Netflix targets a wide range of viewers.
```

```python
#(iii)  Which years saw the highest release of content on Netflix?
```

```python
#code
df['release_year'].value_counts().head(10).plot(kind='bar', color='red')
plt.title("Top Release Years")
plt.show()
```

```python
#INSIGHTS
# Recent years generally have more content, indicating rapid growth of the platform.
# Older content is less frequent, showing focus on modern and trending media.
```

```python
#(iv)  Which countries produce the most Netflix content?
```

```python
#code
countries = df['country'].str.split(', ', expand=True).stack()
countries.value_counts().head(10).plot(kind='bar', color='yellow')
plt.title("Top Countries")
plt.xticks(rotation=45)
plt.show()
```

```python
#INSIGHTS
# The United States is usually the largest contributor of content on Netflix.
# Other countries like India and the UK are also emerging as major content producers.
```

```python
#(v) How has the trend of adding new content evolved year by year?
```

```python
#code
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year

df['year_added'].value_counts().sort_index().plot()

plt.title("Content Added Over Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
#INSIGHTS
#The number of titles added has increased significantly over the years, especially in recent times, 
#showing rapid platform growth and expansion of content.
```

```python
#(vi) Which ratings are most frequent on Netflix?
```

```python
#code
df['rating'].value_counts().head(10).plot(kind='bar', color='violet')
plt.title("Ratings Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
#INSIGHTS
#TV-MA and TV-14 ratings are often the most common, indicating mature audience targeting.
# The friendly content exists but it is comparatively less frequent.
```

```python
#(vii) Do some countries produce more mature content (TV-MA)?
```

```python
#code
mature = df[df['rating'] == 'TV-MA']
mature['country'].value_counts().head(10)
```

```python
#INSIGHTS
#Countries like the United States tend to produce more mature content.
#As this helps in Netflix segment content based on regional audience preferences.
```

```python
#(viii) Which genres are more associated with TV Shows vs Movies?
```

```python
#code
movies = df[df['type'] == 'Movie'].copy()
tv = df[df['type'] == 'TV Show'].copy()
print(movies['listed_in'].value_counts().head())
print(tv['listed_in'].value_counts().head())
```

```python
#INSIGHTS
#Movies are often associated with Drama and Action genres.
#TV Shows commonly focus on series-based genres like TV Dramas and Documentaries.
```

```python
#(ix)  Which genres dominate in the United States vs other countries?
```

```python
#code
us = df[df['country'].str.contains("United States", na=False)]
non_us = df[~df['country'].str.contains("United States", na=False)]
print(us['listed_in'].value_counts().head())
print(non_us['listed_in'].value_counts().head())
```

```python
#INSIGHTS
#The US market often emphasizes mainstream genres like Drama and Comedy.
#Other countries show more diversity, including regional and international genres.
```

```python
#(xv)  What is the average duration of Movies on Netflix?
```

```python
#code
movies = df[df['type'] == 'Movie'].copy()
movies['duration'] = movies['duration'].str.replace(' min', '')
movies['duration'] = pd.to_numeric(movies['duration'], errors='coerce')
print(movies['duration'].mean())
```

```python
#INSIGHTS
#The average movie duration typically falls around standard viewing time (~90–120 minutes).
#This reflects audience preference for moderate-length content.
```

```python
#(xvi) What’s the most common number of seasons for TV shows?
```

```python
#code
tv = df[df['type'] == 'TV Show'].copy()
tv['duration'] = tv['duration'].str.replace(' Seasons', '')
tv['duration'] = tv['duration'].str.replace(' Season', '')
tv['duration'] = pd.to_numeric(tv['duration'], errors='coerce')
tv['duration'].value_counts().head()
```

```python
#INSIGHTS
#Most TV shows have 1–2 seasons, indicating shorter lifecycle.
#Longer series are fewer, suggesting higher production cost or risk.
```

```python
#(xvii) Is there a trend in movie durations over the years?
```

```python
#code
movies.groupby('release_year')['duration'].mean().plot(color='brown')
plt.title("Duration Trend")
plt.show()
```

```python
#INSIGHTS
#Movie durations remain relatively stable over time.
#Slight variations may reflect changing viewer attention spans.
```

```python
#(xviii) In which months does Netflix add the most content?
```

```python
#code
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['month_added'] = df['date_added'].dt.month

df['month_added'].value_counts().sort_index().plot(kind='bar')
plt.show()
```

```python
#INSIGHTS
#Netflix adds more content in mid and later months, showing higher activity towards the end of the year.
```

```python
#(xix) How does the genre distribution vary across different years?
```

```python
#code
df.groupby('year_added')['listed_in'].count().plot(color='purple')
plt.title("Genre Trend")
plt.show()
```

```python
#INSIGHTS
#Genre distribution increases over time with more content diversity.
#Netflix expands its catalog to cater to changing audience interests.
```
